## Visualization & Explanatory Analysis

### Pertanyaan 1: Berapa banyak sepeda yang disewakan pada tahun 2011 selama hari libur ?

```python
# tahun 2011 dan Hari Libur
data = df_day[(df_day["yr"] == 0) & (df_day["holiday"])]

# Hitung jumlah total sewa sepeda
jumlah = data["cnt"].sum()

print("Total sepeda yang disewakan pada tahun 2011 selama hari libur: ", jumlah)
```

### Pertanyaan 2: Apakah terjadi peningkatan jumlah sewa sepeda pada musim 2 dibandingkan dengan musim 1 pada tahun 2011 ?

Musim 2

```python
# tahun 2011 dan season 2
data = df_day[(df_day["yr"] == 0) & (df_day["season"] == 2)]

# Hitung jumlah total sewa sepeda
jumlah = data["cnt"].sum()

print("Jumlah sepeda sewaan yang digunakan pada musim 2 di tahun 2011:", jumlah)
```

Musim 1

```python
# tahun 2011 dan season 2
data_1 = df_day[(df_day["yr"] == 0) & (df_day["season"] == 1)]

# Hitung jumlah total sewa sepeda
jumlah_1 = data_1["cnt"].sum()

print("Jumlah sepeda sewaan yang digunakan pada musim 2 di tahun 2011:", jumlah_1)
```

## Conclusion

- Pada tahun 2011 sebanyak 30022 sepeda telah disewakan selama hari libur.
- Jumlah sepeda sewaan yang digunakan pada musim 2 di tahun 2011 sebanyak 347316, sedangkan pada musim 1 di tahun yang sama Jumlah sepeda sewaan yang digunakan sebanyak 150000. Maka dari terjadi peningkatan pada musim 2 di tahun 2011.
